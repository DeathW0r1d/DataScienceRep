```python
df.drop(['availability','category','delivery_info','description','discount_options','is_favorite','old_category_id','seller','service_prices',
        'sub_category'],axis=1,inplace=True)

df.drop(['seller_type','shark_info','status','tags','recall_info','saleId'],axis=1,inplace=True)

df.drop(['score','services','relevance','additional_info','counters','grouping_info'],axis=1,inplace=True)

df.drop(['hash','id','multiposting','price_history'],axis=1,inplace=True)

df.drop(['soldMinutes','badges','dealer_credit_config'],axis=1,inplace=True)

df.drop(['brand_cert_info','discount_price'],axis=1,inplace=True)

df.drop('lk_summary',axis=1,inplace=True)

df.drop('salon',axis=1,inplace=True)

df['price'] = df.price_info.apply(lambda x: x.get('price',0))

df['mileage'] = df.state.apply(lambda x: x.get('mileage',0))

df['state_not_beaten'] = df.state.apply(lambda x: x.get('state_not_beaten',True))

df['owners_number'] = df.documents.apply(lambda x: x.get('owners_number',0))

df['year'] = df.documents.apply(lambda x: x.get('year'))

df['custom_cleared'] = df.documents.apply(lambda x: x.get('custom_cleared'))

df.drop(['state','documents','price_info'],axis=1,inplace=True)

df['configuration'] = df.vehicle_info.apply(lambda x:x.get('configuration'))

df['equipment'] = df.vehicle_info.apply(lambda x:x.get('equipment'))

df['mark_info'] = df.vehicle_info.apply(lambda x:x.get('mark_info'))

df['model_info'] = df.vehicle_info.apply(lambda x:x.get('model_info'))

df.vehicle_info.iloc[0]['steering_wheel']

df['steering_wheel'] = df.vehicle_info.apply(lambda x:x.get('steering_wheel'))

df.vehicle_info.iloc[0]['tech_param']

df['tech_param'] = df.vehicle_info.apply(lambda x:x.get('tech_param'))

df.vehicle_info.iloc[0]['super_gen']

df['super_gen'] = df.vehicle_info.apply(lambda x:x.get('super_gen'))


```

```python
df.dropna(inplace=True)
```

```python
def body_get(x):
    try:
        res = x.get('body_type')
        return(res)
    except AttributeError:
        return('OTHER')

df['body_type']= df.configuration.apply(lambda x:body_get(x))
```
