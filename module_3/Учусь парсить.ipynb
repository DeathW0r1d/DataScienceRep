```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
browser = webdriver.Chrome(executable_path='chromedriver.exe')
```

```python
city_dict= {'Paris': 568836.0,
 'Stockholm': 61121.0,
 'London': 757656.0,
 'Berlin': 136518.0,
 'Munich': 69446.0,
 'Oporto': 66134.0,
 'Milan': 356036.0,
 'Bratislava': 15116.0,
 'Vienna': 92674.0,
 'Rome': 585930.0,
 'Barcelona': 343737.0,
 'Madrid': 298024.0,
 'Dublin': 100585.0,
 'Brussels': 90397.0,
 'Zurich': 40256.0,
 'Warsaw': 44644.0,
 'Budapest': 121382.0,
 'Copenhagen': 57686.0,
 'Amsterdam': 153559.0,
 'Lyon': 74501.0,
 'Hamburg': 43084.0,
 'Lisbon': 156510.0,
 'Prague': 164318.0,
 'Oslo': 35372.0,
 'Helsinki': 22044.0,
 'Edinburgh': 134954.0,
 'Geneva': 28948.0,
 'Ljubljana': 13141.0,
 'Athens': 59326.0,
 'Luxembourg': 16613.0,
 'Krakow': 50959.0}
```

```python
city_list = list()
population_dict = dict()
density_dict = dict()
area_dict = dict()
for city in city_dict.keys():
    population_dict[city]=0
    density_dict[city]=0
    area_dict[city]=0
    city_list.append(city)
area_dict
```

```python
url_list = list()
for element in city_list:
    if element == 'Luxembourg':
         url_list.append('https://en.wikipedia.org/wiki/Luxembourg_City')
    else:
        url_list.append('https://en.wikipedia.org/wiki/'+str(element))
url_list
```

```python
my_table_list = list()
rows_list = list()
for url in url_list:
    browser.get(url)
    required_HTML = browser.page_source
    soup = BeautifulSoup(required_HTML, 'html5lib')
    table = soup.findChildren('table')
    my_table = table[0]
    my_table_list.append(table[0])
    rows = my_table.findChildren(['th', 'tr'])
    rows_list.append(rows)
```

```python
i = 0
k = 0
parse_list = list()
for rows in rows_list:
    for row in rows:
        cells1 = row.findChildren('th')
        cells2 = row.findChildren('td')
        for cell in cells1:
            value1 = cell.text
            if ('urban' in value1.lower())or('capital city' in value1.lower())or('population' in value1.lower())or('metropolitan city' in value1.lower()):
                k = 1
                parse_list.append([city_list[i],value1,'type 1',k])
            
        for cell in cells2:
            if k>0:
                value2 = cell.text
                parse_list.append([city_list[i],value2,'type 2',k])
                k += -1
    i += 1
parse_list
```

Составим словарь, ключами которого будет являться город, а значениями- его население:

```python

for element in parse_list:
    if (not('km2'in element[1]))and(not('Urban' in element[1]))and(not('%') in element[1])and(not(' by ') in element[1]):
        population_dict[element[0]] = element[1]
population_dict
```

Необходимо почистить данные по населению:

```python
import re
```

```python
for city,population in population_dict.items():
    population_dict[city]= (re.findall('((\d+\,{0,1}){1,3})',str(population)))[0][0]                      
for city,population in population_dict.items():
    population_dict[city]= int(population.replace(',',''))
population_dict
```

```python
i = 0
k = 0
parse_list_d = list()
for rows in rows_list:
    for row in rows:
        cells1 = row.findChildren('th')
        cells2 = row.findChildren('td')
        for cell in cells1:
            value1 = cell.text
            if ('density' in value1.lower())and(not('metro' in value1.lower())):
                k = 1
                parse_list_d.append([city_list[i],value1,'type 1',k])
            
        for cell in cells2:
            if k>0:
                value2 = cell.text
                parse_list_d.append([city_list[i],value2,'type 2',k])
                k += -1
    i += 1
parse_list_d
```

```python
for element in parse_list_d:
    if ('km2'in element[1]):
        density_dict[element[0]] = element[1]
density_dict
```

```python
for city,population in density_dict.items():
    density_dict[city]= (re.findall('((\d+\,{0,1}){1,3})',str(population)))[0][0]                      

for city,population in density_dict.items():
    density_dict[city]= int(population.replace(',',''))
density_dict
```

```python
i = 0
k = 0
parse_list_a = list()
for rows in rows_list:
    for row in rows:
        cells1 = row.findChildren('th')
        cells2 = row.findChildren('td')
        for cell in cells1:
            value1 = cell.text
            if ('area' in value1.lower()):
                k = 1
                parse_list_a.append([city_list[i],value1,'type 1',k])
            
        for cell in cells2:
            value2 = cell.text
            if k>0 and ('km2' in value2.lower()):
                value2 = cell.text
                parse_list_a.append([city_list[i],value2,'type 2',k])
                k += -1
    i += 1
parse_list_a
```

```python
for element in parse_list_a:
    if ('km2'in element[1]):
        area_dict[element[0]] = element[1]
area_dict
```
